# Data Uploading

```python
import pandas as pd
import math
luxuryhouse_df = pd.read_csv('/content/Luxury_Housing_Bangalore.csv')
luxuryhouse_df.head()
```

# renaming the column

```python
luxuryhouse_df = luxuryhouse_df.rename(columns={'Micro_Market':'Area','Configuration':'Bedrooms'})
luxuryhouse_df['Ticket_Price_Cr'] = pd.to_numeric(luxuryhouse_df['Ticket_Price_Cr'], errors='coerce')
luxuryhouse_df.head()
```

# Drop Duplicates

```python
luxuryhouse_df.info()
```

```python
luxuryhouse_df['Property_ID'] = luxuryhouse_df['Property_ID'].astype(str).str.replace(r'\D', '', regex=True)
luxuryhouse_df['Project_Name'] = luxuryhouse_df['Project_Name'].astype(str).str.replace(r'\D', '', regex=True)
luxuryhouse_df['Bedrooms'] = luxuryhouse_df['Bedrooms'].astype(str).str.replace(r'\D', '', regex=True)

luxuryhouse_df.tail()
```

```python
luxuryhouse_df['Bedrooms'] = pd.to_numeric(luxuryhouse_df['Bedrooms'],errors='coerce')
luxuryhouse_df.info()
```

# Duplicates


```python
luxuryhouse_df[luxuryhouse_df.duplicated].reset_index(drop=True)
luxuryhouse_df.drop_duplicates(inplace=True)
luxuryhouse_df.info()
```

```python
luxuryhouse_df['Ticket_Price_Cr'] = np.where(luxuryhouse_df['Ticket_Price_Cr'].isnull(),luxuryhouse_df['Ticket_Price_Cr'].mode(),luxuryhouse_df['Ticket_Price_Cr'])
```

```python
luxuryhouse_df.isnull().sum()
```

```python
import numpy as np

# Calculate the mode of the 'Amenity_Score' column
mode_value = luxuryhouse_df['Amenity_Score'].mode()[0]

# Fill nulls with the mode
luxuryhouse_df['Amenity_Score'] = luxuryhouse_df['Amenity_Score'].fillna(mode_value)
```
